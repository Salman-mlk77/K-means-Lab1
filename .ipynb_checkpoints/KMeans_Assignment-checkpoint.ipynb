# EDA

## Create a scatterplot of Grad.Rate versus Room.Board where the points are colored by the Private column. 

```python
plt.figure(figsize=(14, 8))
sns.scatterplot(data=df, x='Room.Board', y='Grad.Rate', hue='Private', s=80, edgecolor='none')
plt.grid()
plt.title('Scatterplot of Grad.Rate versus Room.Board')
plt.legend(loc='center right');
```

## Create a scatterplot of F.Undergrad versus Outstate where the points are colored by the Private column.

```python
plt.figure(figsize=(14, 8))
sns.scatterplot(data = df, x='Outstate', y='F.Undergrad' , hue='Private', alpha=0.4, s=80, edgecolor='none');
plt.grid()
plt.title('Scatterplot of F.Undergrad versus Outstate')
plt.legend(loc='center right');
```

## Use sns.FacetGrid to create a stacked histogram showing Out of State Tuition based on the Private column. 

```python
 sns.set_style("darkgrid")
g = sns.FacetGrid(data=df, hue='Private', height=6, aspect=2.5)
g.map(sns.histplot, 'Outstate',bins=20, stat="count", alpha=0.3);
```

## Use sns.FacetGrid to create a stacked histogram showing Grad.Rate based on the Private column. 

```python
sns.set_style("darkgrid")
g = sns.FacetGrid(df, hue="Private", height=6 , aspect=2.5)
g.map(sns.histplot, 'Grad.Rate', stat="count",bins=20, alpha=0.3);
```

## Notice how there seems to be a private school with a graduation rate of higher than 100%.What is the name of that school?

```python
# To print school name
school_name = df[(df['Private'] == 'Yes') & (df['Grad.Rate'] > 100)]['Private'].index[0]
print(school_name)
```

## K Means Model

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

## Fit the model to all the data except for the Private label.

```python
X = df.drop("Private",axis=1)
```

```python
prediction = kmeans.fit_predict(X)
```

## What are the cluster center vectors?

```python
kmeans.cluster_centers_
```

## Create a new column for df called 'Cluster', which is a 1 for a Private school, and a 0 for a public school.

```python
df['Cluster'] = np.where(df['Private'] == 'Yes', 1, 0)
```

```python
df.head()
```

## Create a confusion matrix and classification report to see how well the Kmeans clustering worked without being given any labels.

```python
from sklearn.metrics import confusion_matrix

y = df['Cluster']
confusion_matrixx = confusion_matrix(y,prediction)
print(confusion_matrixx)
```

```python
from sklearn.metrics import classification_report

class_report = classification_report(y,prediction)
print(class_report)
```
